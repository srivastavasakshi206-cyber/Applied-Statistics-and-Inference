Q1.  What are Type I and Type II errors in hypothesis testing, and how do they
imapact decision-making?
ANS. In hypothesis testing, we aim to determine if there is enough evidence to reject a null hypothesis ($H_0$) in favor of an alternative hypothesis ($H_a$). However, there is always a risk of making an incorrect decision. These incorrect decisions are classified as Type I and Type II errors.

**Type I Error (False Positive):**A Type I error occurs when you reject the null hypothesis ($H_0$) when it is actually true.
*   **Impact on Decision-Making:** A Type I error can lead to unnecessary actions or conclusions based on a false premise. For example, in a clinical trial, a Type I error could lead to approving a drug that is not effective or has harmful side effects.

**Type II Error (False Negative):**A Type II error occurs when you fail to reject the null hypothesis ($H_0$) when the alternative hypothesis ($H_a$) is actually true.
*   **Impact on Decision-Making:** A Type II error can lead to missing important findings or opportunities. For example, in a clinical trial, a Type II error could lead to failing to approve a drug that is actually effective and could benefit patients.

**Impact on Decision-Making Summary:**
*   **Type I errors** lead to acting when you shouldn't (false alarm).
*   **Type II errors** lead to not acting when you should (missed opportunity).

Q2. What is the P-value in hypothesis testing, and how should it be interpreted in the context of the null hypothesis?
ANS. In hypothesis testing, the P-value (or probability value) is a crucial piece of information that helps us decide whether to reject the null hypothesis ($H_0$).

**Definition of P-value:**
The P-value is the probability of obtaining test results at least as extreme as the observed results, assuming that the null hypothesis ($H_0$) is true. In simpler terms, it's the probability of seeing the data you have, or something even more unlikely, if there is actually no effect or no difference (as stated by the null hypothesis).

**Interpretation of the P-value:**

The P-value is compared to the significance level ($\alpha$), which you set before conducting the hypothesis test. Common values for $\alpha$ are 0.05, 0.01, or 0.10.

*   **If P-value $\le \alpha$:** This is considered strong evidence against the null hypothesis. You would **reject the null hypothesis**. This suggests that the observed results are unlikely to have occurred by random chance alone if the null hypothesis were true.
*   **If P-value $> \alpha$:** This is considered weak evidence against the null hypothesis. You would **fail to reject the null hypothesis**. This suggests that the observed results are reasonably likely to have occurred by random chance if the null hypothesis were true. It's important to note that failing to reject the null hypothesis does **not** mean you accept the null hypothesis as true. It simply means you don't have enough evidence to reject it.

Q3. Explain the difference between a Z-test and a T-test, including when to use each.
ANS. Z-tests and T-tests are both statistical hypothesis tests used to determine if there is a significant difference between the means of two groups or between a sample mean and a hypothesized population mean. The key difference lies in the information you have about the population standard deviation.

**Z-test:**A Z-test is used when you know the population standard deviation ($\sigma$)**. This is a strong assumption and is often not the case in real-world scenarios.
*  The Z-test statistic follows a standard normal distribution (Z-distribution).

**T-test:**A T-test is used when you do not know the population standard deviation ($\sigma$)** and you have to estimate it using the sample standard deviation (s). This is a much more common scenario.The data is approximately normally distributed (especially for small sample sizes).
* The T-test statistic follows a t-distribution with $n-1$ degrees of freedom. The t-distribution is similar to the normal distribution but has heavier tails, especially for small sample sizes, accounting for the uncertainty introduced by using the sample standard deviation to estimate the population standard deviation. As the sample size increases, the t-distribution approaches the normal distribution.

In practice, T-tests are used more frequently than Z-tests because the population standard deviation is rarely known.

Q4. What is a confidence interval, and how does the margin of error influence
its width and interpretation?
ANS. In statistics, a confidence interval is a range of values that is likely to contain a population parameter with a certain level of confidence. It provides a range of plausible values for an unknown population parameter (like a population mean or proportion) based on a sample from that population.

* Purpose:Confidence intervals are used to estimate an unknown population parameter. Because we usually only have a sample of data from a population, we can't know the exact value of the population parameter. A confidence interval gives us a range within which we are confident the true population parameter lies.
* Structure:A confidence interval is typically expressed as:

    Estimate $\pm$ Margin of Error

    Where:
    *   **Estimate:** This is the point estimate of the population parameter, usually calculated from the sample (e.g., sample mean $\bar{x}$ for estimating the population mean $\mu$, or sample proportion $\hat{p}$ for estimating the population proportion $p$).
    *   **Margin of Error:** This accounts for the uncertainty in our estimate due to sampling variability. It represents the maximum likely difference between the sample estimate and the true population parameter.

*   **Confidence Level:** A confidence interval is associated with a confidence level, often expressed as a percentage (e.g., 95% confidence level). This percentage represents the proportion of times that the confidence interval constructed from repeated random samples would contain the true population parameter. It does **not** mean there is a 95% probability that the true parameter is within the specific interval you calculated.

**How the Margin of Error Influences Width and Interpretation:**

The margin of error is a critical component of the confidence interval and directly influences its width and our interpretation of the results.

*   **Influence on Width:** The confidence interval's width is twice the margin of error (Width = 2 * Margin of Error).
    *   **Larger Margin of Error = Wider Confidence Interval:** A larger margin of error results in a wider confidence interval. This means we are less precise in our estimate of the population parameter, but we have a higher confidence that the true parameter is within that wider range.
    *   **Smaller Margin of Error = Narrower Confidence Interval:** A smaller margin of error results in a narrower confidence interval. This means we are more precise in our estimate, but our confidence level might be lower (for a fixed sample size).

*   **Factors Affecting the Margin of Error:** The margin of error is influenced by several factors:
    *   **Confidence Level:** A higher confidence level (e.g., 99% vs. 95%) requires a larger margin of error to be more certain that the interval contains the true parameter.
    *   **Sample Size:** A larger sample size reduces the variability of the sample estimate, leading to a smaller margin of error and a narrower confidence interval.
    *   **Variability of the Data (Standard Deviation):** Higher variability in the population (larger standard deviation) leads to a larger margin of error and a wider confidence interval.

Q5. Describe the purpose and assumptions of an ANOVA test. How does it extend hypothesis testing to more than two groups?
ANS. ANOVA, which stands for Analysis of Variance, is a statistical test used to determine if there are any statistically significant differences between the means of three or more independent groups. It is a powerful tool for comparing means when you have more than two groups, as conducting multiple t-tests increases the risk of making a Type I error.

Purpose of ANOVA:
The primary purpose of ANOVA is to test whether there is a significant difference between the means of multiple groups. Instead of directly comparing group means, ANOVA analyzes the variance within and between the groups.

ANOVA calculates an F-statistic, which is the ratio of the variance between groups to the variance within groups.

*Variance Between Groups:This measures the variability of the group means around the overall mean of the data. A larger between-group variance suggests that the group means are further apart.
*Variance Within Groups:This measures the variability of the data points within each group around their respective group mean. A smaller within-group variance suggests that the data points within each group are clustered closely together.

If the variance between groups is significantly larger than the variance within groups, it suggests that the differences between the group means are unlikely to be due to random chance, and we can reject the null hypothesis.

**Assumptions of ANOVA:**

For the results of an ANOVA test to be valid, several assumptions should be met:

*Independence of Observations:The observations within and between groups must be independent. This means that the value of one observation does not influence the value of another.
*Normality:The data within each group should be approximately normally distributed. ANOVA is relatively robust to violations of this assumption, especially with larger sample sizes, but significant deviations from normality can affect the results.
*Homogeneity of Variances (Homoscedasticity):The variances of the data in each group should be approximately equal. If the variances are significantly different between groups (heteroscedasticity), it can affect the F-statistic and the p-value. Levene's test or Bartlett's test can be used to check this assumption.

Q6.  Write a Python program to perform a one-sample Z-test and interpret the
result for a given dataset.

```python
import numpy as np
from statsmodels.stats.weightstats import DescrStatsW

data = [175, 168, 180, 170, 172, 178, 176, 173, 171, 179]

hypothesized_mean = 170

population_std_dev = 5

alpha = 0.05

dstat = DescrStatsW(data, ddof=0)
z_statistic, p_value = dstat.ztest_mean(value=hypothesized_mean, alternative='two-sided')

print(f"One-Sample Z-Test Results:")
print(f"  Hypothesized Mean: {hypothesized_mean}")
print(f"  Known Population Standard Deviation: {population_std_dev}")
print(f"  Z-Statistic: {z_statistic:.4f}")
print(f"  P-Value: {p_value:.4f}")
print(f"  Significance Level (alpha): {alpha}")

if p_value <= alpha:
    print("\nDecision: Reject the null hypothesis.")
    print(f"Interpretation: There is sufficient evidence to conclude that the true population mean is significantly different from {hypothesized_mean} at the {alpha} significance level.")
else:
    print("\nDecision: Fail to reject the null hypothesis.")
    print(f"Interpretation: There is not enough evidence to conclude that the true population mean is significantly different from {hypothesized_mean} at the {alpha} significance level.")

sample_mean = np.mean(data)
sample_std_dev = np.std(data, ddof=1)

print(f"\nSample Mean: {sample_mean:.2f}")
print(f"Sample Standard Deviation: {sample_std_dev:.2f}")
```

Q7.  Simulate a dataset from a binomial distribution (n = 10, p = 0.5) using

Q9.  Write a Python function to calculate and visualize the confidence interval
for a sample mean.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

def calculate_and_visualize_confidence_interval(data, confidence_level=0.95):
    """
    Calculates and visualizes the confidence interval for a sample mean.

    Args:
        data (array-like): The sample data.
        confidence_level (float): The desired confidence level (e.g., 0.95 for 95%).
    """
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)
    sample_size = len(data)

    standard_error = sample_std / np.sqrt(sample_size)

    degrees_of_freedom = sample_size - 1
    t_score = stats.t.ppf((1 + confidence_level) / 2, degrees_of_freedom)

    margin_of_error = t_score * standard_error

    confidence_interval_lower = sample_mean - margin_of_error
    confidence_interval_upper = sample_mean + margin_of_error

    print(f"Sample Mean: {sample_mean:.4f}")
    print(f"Sample Standard Deviation: {sample_std:.4f}")
    print(f"Sample Size: {sample_size}")
    print(f"Confidence Level: {confidence_level}")
    print(f"Degrees of Freedom: {degrees_of_freedom}")
    print(f"T-Score: {t_score:.4f}")
    print(f"Margin of Error: {margin_of_error:.4f}")
    print(f"Confidence Interval: ({confidence_interval_lower:.4f}, {confidence_interval_upper:.4f})")

    plt.figure(figsize=(8, 4))
    plt.errorbar(sample_mean, 0, xerr=margin_of_error, fmt='o', capsize=5, markersize=8, label=f'{int(confidence_level*100)}% Confidence Interval')
    plt.scatter(sample_mean, 0, color='red', zorder=5, label='Sample Mean')
    plt.title("Confidence Interval for the Sample Mean")
    plt.xlabel("Value")
    plt.yticks([])
    plt.legend()
    plt.grid(axis='x', alpha=0.5)
    plt.show()

np.random.seed(42)
sample_data = np.random.normal(loc=50, scale=10, size=30)

calculate_and_visualize_confidence_interval(sample_data, confidence_level=0.95)
```

Q10.  Perform a Chi-square goodness-of-fit test using Python to compare
observed and expected distributions, and explain the outcome

```python
import numpy as np
from scipy.stats import chisquare

observed_frequencies = np.array([20, 25, 15, 22, 18, 20])

total_rolls = np.sum(observed_frequencies)
expected_frequencies = np.array([total_rolls / 6] * 6)

print(f"Observed Frequencies: {observed_frequencies}")
print(f"Expected Frequencies: {expected_frequencies}")

chi_square_statistic, p_value = chisquare(f_obs=observed_frequencies, f_exp=expected_frequencies)

alpha = 0.05

print(f"\nChi-square Statistic: {chi_square_statistic:.4f}")
print(f"P-value: {p_value:.4f}")
print(f"Significance Level (alpha): {alpha}")

print("\nInterpretation:")
if p_value <= alpha:
    print(f"Since the P-value ({p_value:.4f}) is less than or equal to the significance level ({alpha}), we reject the null hypothesis.")
    print("This suggests that there is a statistically significant difference between the observed and expected distributions.")
    print("In the context of the die roll example, there is evidence to suggest that the die is not fair.")
else:
    print(f"Since the P-value ({p_value:.4f}) is greater than the significance level ({alpha}), we fail to reject the null hypothesis.")
    print("This suggests that there is not enough evidence to conclude that there is a statistically significant difference between the observed and expected distributions.")
    print("In the context of the die roll example, there is not enough evidence to suggest that the die is not fair.")
```

**Explanation of the Outcome:**

The Chi-square goodness-of-fit test assesses how well an observed distribution matches an expected theoretical distribution.

*   **Null Hypothesis ($H_0$):** The observed distribution is the same as the expected distribution (i.e., there is no significant difference).
*   **Alternative Hypothesis ($H_a$):** The observed distribution is different from the expected distribution.

The test calculates a **Chi-square statistic**, which measures the discrepancy between the observed and expected frequencies. A larger Chi-square statistic indicates a greater difference between the observed and expected distributions.

The **P-value** is the probability of obtaining a Chi-square statistic as extreme as, or more extreme than, the one calculated from the sample data, assuming the null hypothesis is true.

*   If the P-value is small (typically $\le \alpha$), it means the observed data is unlikely to have occurred if the null hypothesis were true. Therefore, we reject the null hypothesis and conclude that the observed distribution is significantly different from the expected distribution.
*   If the P-value is large (typically $> \alpha$), it means the observed data is reasonably likely to have occurred if the null hypothesis were true. Therefore, we fail to reject the null hypothesis, indicating that there is not enough evidence to conclude a significant difference between the observed and expected distributions.

In the example above, we compared the observed frequencies of die rolls to the expected frequencies of a fair die. Based on the calculated P-value and the chosen significance level, we make a decision about whether to reject the null hypothesis (that the die is fair) or fail to reject it.
